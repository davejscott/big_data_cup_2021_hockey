```python
def change_clock_time(clock_time):
    """
    take str clock
    format and return as time in seconds
    *** will update later to include period, didn't do now to avoid
    issues with added time in overtime
    """
    x = time.strptime(clock_time.split(',')[0],'%M:%S')
    time_in_seconds = datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds()
    return time_in_seconds
```

## Create a filtered dataframe that only contains passes (Play), Shots (Shot), and Goals(Goal)

```python
df_filter_plays = pd.concat([pd.concat([df.loc[df.Event == 'Play'], df.loc[df.Event == 'Shot']], axis=0), df.loc[df.Event == 'Goal']], axis = 0)
df_filter_plays.head(5)
```

```python
filter_df_columns = [x for x in df.columns]
filter_df_columns.append('Time Between Shot')
filter_df_columns.append('Previous Pass Index')
```

# Using this to create a stream where we look on a play where
## a puck has been accepted by the original action (pass) which is Player 2 then it is shot by Player 2 of the pass (who is now Player) in the secondary action
## action: is the original action (right now looking for pass)
## action_filtered is the secondary action (right now is a shot or goal)
## The goal of this filtering is to create a dataframe where a player has accepted a pass and then is shot or scores a goal within 30 seconds of recieving that pass.
## We also track the time between shot and receiving pass, as well as the previous index of that pass

```python
new_data = pd.DataFrame(columns = filter_df_columns)
for og_index, action in df_filter_plays.iterrows():
    if action['Event'] == "Play":
        print(og_index)
        for index_filtered, action_filtered in df_filter_plays.iterrows():
            if index_filtered < og_index:
                continue
            else:
                time_difference = action['Clock'] - action_filtered['Clock'] 
                if time_difference > 30:
                    break
                if action_filtered['Player'] == action['Player 2']:
                    if action_filtered['Event'] == 'Shot' or action_filtered['Event'] == 'Goal':
                        action_to_append = action_filtered.append(pd.Series(time_difference, index=['Time Between Shot']))
                        action_to_append = action_to_append.append(pd.Series(og_index, index = ['Previous Pass Index']))
                        new_data = new_data.append(pd.DataFrame(action_to_append).T)
                        break
                    else:
                        continue
    else:
        continue
    if og_index > 5000:
        break
```
